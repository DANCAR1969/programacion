Clasificaci√≥n de flores con Machine Learning Supervisado  (Iris Dataset) utilizando el ![image.png]

```python
# Paso 1: Instalar librer√≠as necesarias (ya incluidas en Google Colab)
# No necesitas instalar scikit-learn o pandas manualmente, pero si deseas:
# !pip install scikit-learn pandas matplotlib

# Paso 2: Importar las librer√≠as necesarias
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Paso 3: Cargar el dataset Iris
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.Series(iris.target)

# Paso 4: Dividir los datos en conjunto de entrenamiento y prueba (70% - 30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Paso 5: Crear el modelo de clasificaci√≥n (√Årbol de Decisi√≥n)
modelo = DecisionTreeClassifier(random_state=42)
modelo.fit(X_train, y_train)

# Paso 6: Realizar predicciones
y_pred = modelo.predict(X_test)

# Paso 7: Evaluar el modelo
print("üîç Reporte de Clasificaci√≥n:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))

# Paso 8: Matriz de Confusi√≥n
matriz_conf = confusion_matrix(y_test, y_pred)

# Visualizar la matriz
plt.figure(figsize=(6,4))
sns.heatmap(matriz_conf, annot=True, cmap="Blues", xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.title("Matriz de Confusi√≥n")
plt.xlabel("Predicci√≥n")
plt.ylabel("Valor Real")
plt.show()
```

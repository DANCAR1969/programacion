```python
!pip install gTTS

from gtts import gTTS  # Importa gTTS
from IPython.display import Audio  # Para reproducir el archivo de audio

def speak_and_print(text, filename="audio.mp3"):
    """
    Función para imprimir el texto y generar el audio usando gTTS.
    Luego, reproduce el archivo generado en Google Colab.
    """
    print(text)  # Muestra el texto en la consola
    tts = gTTS(text=text, lang='es')  # Convierte el texto a voz (en español)
    tts.save(filename)  # Guarda el archivo de audio
    display(Audio(filename, autoplay=True))  # Reproduce el archivo en Colab

# Solicita datos al usuario
marca = input("¿Cuál es la marca de tu vehículo? ")
speak_and_print(f"{marca} es la marca más vendida del mercado.")

serie = input("¿Cuál es la serie de tu vehículo? ")
speak_and_print(f"{serie} es la versión más confortable.")

# Maneja el cilindraje con validación
try:
    potencia = int(input(f"¿Qué cilindraje tiene tu {marca}? "))

    if potencia == 3500:
        speak_and_print(f"Tu {marca} {serie} es el de mayor potencia.")
    elif potencia >= 2000:
        speak_and_print(f"Tu {marca} {serie} es de potencia robusta.")
    else:
        speak_and_print(f"Tu {marca} {serie} es de potencia estándar.")
except ValueError:
    speak_and_print("Error: El cilindraje debe ser un número entero.")
```

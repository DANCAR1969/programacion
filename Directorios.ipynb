En Python, los directorios son una estructura que organiza archivos y carpetas en un sistema de archivos. Se utilizan para almacenar y organizar proyectos, bibliotecas, módulos y otros recursos. A continuación, se explica detalladamente cómo funcionan los directorios en Python:

1. Directorio Actual (Current Directory)
El directorio actual es donde se ejecuta un script de Python. Puedes obtenerlo y cambiarlo utilizando el módulo os:

```python
import os

# Obtener el directorio actual
directorio_actual = os.getcwd()
print("Directorio actual:", directorio_actual)

# Crear una nueva carpeta dentro del entorno de Colab
nuevo_directorio = "/content/nueva_carpeta"
os.makedirs(nuevo_directorio, exist_ok=True)

# Cambiar el directorio actual a la nueva carpeta
os.chdir(nuevo_directorio)
print("Directorio cambiado a:", os.getcwd())
```

2. Crear y Eliminar Directorios

Puedes crear y eliminar directorios usando os:

```python
# Crear un directorio
os.mkdir("nuevo_directorio")

# Crear múltiples directorios anidados
os.makedirs("directorio/principal/secundario")

# Eliminar un directorio
os.rmdir("nuevo_directorio")

# Eliminar múltiples directorios
os.removedirs("directorio/principal/secundario")
```

3. Listar Archivos y Directorios

Para listar archivos y subdirectorios en una carpeta, utiliza os.listdir():

```python
contenido = os.listdir(".")
print("Contenido del directorio actual:", contenido)
```

4. Navegar y Explorar Directorios

Puedes explorar recursivamente un árbol de directorios usando os.walk():

```python
for raiz, directorios, archivos in os.walk("."):
    print(f"Directorio actual: {raiz}")
    print(f"Subdirectorios: {directorios}")
    print(f"Archivos: {archivos}")
```

5. Operaciones con Rutas de Archivos

El módulo os.path permite trabajar con rutas:

```python
# Unir rutas
ruta_completa = os.path.join("directorio", "archivo.txt")

# Comprobar si una ruta existe
existe = os.path.exists(ruta_completa)
print(f"¿La ruta existe? {existe}")

# Comprobar si es archivo o directorio
es_archivo = os.path.isfile(ruta_completa)
es_directorio = os.path.isdir("directorio")
print(f"¿Es archivo? {es_archivo}, ¿Es directorio? {es_directorio}")
```

6. Uso de pathlib (Alternativa Moderna)

Desde Python 3.4, se recomienda usar pathlib, que proporciona una forma más intuitiva de trabajar con directorios y archivos:

```python
from pathlib import Path

# Crear una ruta
directorio = Path("nueva_carpeta")

# Crear el directorio
directorio.mkdir(parents=True, exist_ok=True)

# Listar contenido
for archivo in directorio.iterdir():
    print(archivo)

# Comprobar si existe
if directorio.exists():
    print("El directorio existe")
```

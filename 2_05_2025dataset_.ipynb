```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset
ruta = "/content/drive/MyDrive/kevin/Automobile_data.csv"
df = pd.read_csv(ruta)

# Convertir precio a numérico
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df = df.dropna(subset=['price'])

# Agrupar por marca: contar cuántas veces aparece y obtener el precio promedio
ventas_por_marca = df.groupby('make').agg({
    'make': 'count',
    'price': 'mean'
}).rename(columns={'make': 'cantidad_vendida', 'price': 'precio_promedio'})

# Ordenar por cantidad vendida y tomar las top 10
top_ventas = ventas_por_marca.sort_values('cantidad_vendida', ascending=False).head(10)

# Crear gráfico de barras
plt.figure(figsize=(12, 6))
plt.bar(top_ventas.index, top_ventas['cantidad_vendida'], color='steelblue', label='Cantidad Vendida')
plt.ylabel('Cantidad Vendida')
plt.xlabel('Marca')
plt.title('10 Marcas Más Vendidas y Precio Promedio')
plt.xticks(rotation=45)

# Agregar una segunda escala para el precio promedio
plt.twinx()
plt.plot(top_ventas.index, top_ventas['precio_promedio'], color='orange', marker='o', label='Precio Promedio')
plt.ylabel('Precio Promedio ($)')

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ruta del archivo
ruta = "/content/drive/MyDrive/kevin/Dataset_Empresas_de_Compra-Venta_de_Veh_culos_Usados_en_el_Dpto._del_Huila_20250502.csv"

# Cargar el dataset
df = pd.read_csv(ruta)

# Limpiar nombres de columnas
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace('á', 'a')
    .str.replace('é', 'e')
    .str.replace('í', 'i')
    .str.replace('ó', 'o')
    .str.replace('ú', 'u')
)

# Mostrar columnas disponibles
print("Columnas disponibles:", df.columns)

# Seleccionar la primera columna categórica (tipo 'object') y la primera numérica
col_cat = df.select_dtypes(include='object').columns[0]
col_num = df.select_dtypes(include='number').columns[0]

# Agrupar por la columna categórica y sumar la numérica
df_grouped = df.groupby(col_cat)[col_num].sum().sort_values(ascending=False).head(10)  # top 10

# Crear la gráfica de barras
plt.figure(figsize=(12, 6))
plt.bar(df_grouped.index, df_grouped.values, color='teal')
plt.xlabel(col_cat.capitalize())
plt.ylabel(col_num.capitalize())
plt.title(f"{col_num.capitalize()} total por {col_cat.capitalize()}")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

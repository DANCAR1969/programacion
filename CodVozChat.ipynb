```python
# Maria Claudia Raga - Inteligencia Artificial IAP2-140
# Algoritmo para interactuar texto y voz como un chat - voz

# Se instala libreria que permite usar Voz
!pip install gTTS
from gtts import gTTS
from IPython.display import Audio, display
import tempfile
import time # Importing the 'time' module

# Definimos la funcion que convierte el texto en audio
def hablar(texto):
    tts = gTTS(text=texto, lang='es')
    with tempfile.NamedTemporaryFile(delete=False, suffix=".mp3") as fp:
        tts.save(fp.name)
        display(Audio(fp.name, autoplay=True))
        time.sleep(1)

# Definimos las variables y la funcion con la que iniciamos el programa
def chat_por_voz():
    respuestas = []
    cad=[]
    resumen = ""
    hablar("Hola Bienvenido a nuestra guia turistica.")
    time.sleep(3)

# Preguntas que con las que interactuamos con el usuario
    preguntas = [
        "¿Cual es tu nombre?",
        "¿En que ciudad te encuentras ?",
        "¿Que tipo de actividades te interesan (actividad aire libre, nocturnas, museos)?",
        "¿Cual es tu presupuesto?"
    ]
# Ciclo para realizar preguntas y obtener repuesta de cada pregunta
# Cada respuesta se guarda como una lista o cadena
    for pregunta in preguntas:
        print(pregunta)
        hablar(pregunta)
        respuesta = input("Ingresa tu Respuesta: ").upper()
        cad.append(respuesta)

#Elegimos la recomendacion de acuerdo a las principales ciudades que pudo ingresar el cliente

    if cad[1]=="CALI":
        recom="la tertulia"
    elif cad[1]=="MEDELLIN":
        recom="el pueblito paisa"
    elif cad[1]=="BOGOTA":
        recom="el cerro de Monserrate"
    else:
        recom="No tengo ninguna recomendacion para la ciudad donde estas"

# Mensaje final con la informacion que ingreso el usuario
    resumen += f"Hola {cad[0]}', te damos la bienvenida a la ciudad de {cad[1]}, super que disfrutes {cad[2]} y como cuentas con un presupuesto {cad[3]} para el dia de hoy te invitamos a conocer {recom}"
    print(resumen)
    hablar(resumen)

# Ejecutar la funcion para iniciar el programa
chat_por_voz()
```

El objetivo del codigo es Predecir el precio de una casa basado en su tamaño (en metros cuadrados) usando un modelo de regresión lineal supervisada.

```python
# Paso 1: Importar librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Paso 2: Crear los datos de entrenamiento (tamaño de casa en m² y su precio)
X = np.array([[50], [80], [100], [120], [150]])  # Tamaño en m²
y = np.array([100000, 160000, 200000, 240000, 300000])  # Precio en dólares

# Paso 3: Crear el modelo de regresión lineal
modelo = LinearRegression()

# Paso 4: Entrenar el modelo con los datos
modelo.fit(X, y)

# Paso 5: Realizar una predicción para una casa de 110 m²
tamaño_nuevo = np.array([[110]])
prediccion = modelo.predict(tamaño_nuevo)

# Mostrar el resultado
print(f"Precio estimado para una casa de 110 m²: ${prediccion[0]:,.2f}")

# Paso 6: Visualizar los datos y la regresión
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, modelo.predict(X), color='red', label='Línea de regresión')
plt.scatter(tamaño_nuevo, prediccion, color='green', label='Predicción 110 m²')
plt.xlabel("Tamaño de la casa (m²)")
plt.ylabel("Precio (USD)")
plt.title("Regresión Supervisada: Tamaño vs Precio")
plt.legend()
plt.grid(True)
plt.show()
```

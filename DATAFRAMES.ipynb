**DataFrames en Python: **

**Definición**
Un DataFrame es una estructura de datos bidimensional similar a una tabla en una hoja de cálculo, base de datos o archivo CSV. Forma parte de la biblioteca Pandas en Python y permite almacenar y manipular datos de manera eficiente.


Sintaxis Básica:

```python
import pandas as pd


datos = {'Nombre': ['Ana', 'Luis', 'Pedro'], 'Edad': [25, 30, 22], 'Ciudad': ['Madrid', 'Barcelona', 'Valencia']}
print(datos)
df = pd.DataFrame(datos)

print(df)
```

**Tipos de DataFrames en Python**

- DataFrame Simple: Tablas de datos sencillas con Pandas.
- DataFrame Multinivel: Estructura jerárquica con índices múltiples.
- DataFrame con Series: Combinación de varias Series en un DataFrame.

Ejemplos detallados de algunos tipos de DataFrames en Python:

1. Pandas DataFrame (Estándar)
El DataFrame más utilizado en Python, ideal para manipulación y análisis de datos.

```python
import pandas as pd

# Crear un DataFrame
datos = {
    'Nombre': ['Ana', 'Luis', 'Pedro'],
    'Edad': [25, 30, 22],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia']
}
df = pd.DataFrame(datos)

# Mostrar el DataFrame
print(df)
```

2. Dask DataFrame (Datos Grandes)

Se utiliza para manejar grandes volúmenes de datos en paralelo.

```python

import dask.array as da
# Create a large Dask array
x = da.random.random((10, 10), chunks=(10, 10))
print('Gneerated Input')
print(x.compute())
# Perform a computation
result = x.mean().compute()
print('Gneerated Mean')
print(result)
```

3. Koalas DataFrame (Big Data con Apache Spark)

Permite usar una sintaxis similar a Pandas para trabajar con PySpark.

```python
# Instalar PySpark e importar librerías
!pip install -q pyspark
from pyspark.sql import SparkSession
from pyspark.sql import Row

# Crear la sesión de Spark
spark = SparkSession.builder.appName("EjemploSimpleSpark").getOrCreate()

# Crear datos como objetos Row
datos = [
    Row(Nombre="Ana", Edad=25),
    Row(Nombre="Luis", Edad=30),
    Row(Nombre="Pedro", Edad=22)
]

# Crear DataFrame y mostrarlo
df = spark.createDataFrame(datos)
df.show()
```

4. GeoPandas DataFrame (Datos Geoespaciales)

Ideal para análisis de datos espaciales y mapas.

```python
# Paso 1: Instalar geopandas y shapely
!pip install -q geopandas shapely pyproj fiona

# Paso 2: Importar las librerías necesarias
import geopandas as gpd
from shapely.geometry import Point

# Paso 3: Crear un diccionario con coordenadas y ciudades
data = {
    'Ciudad': ['Bogotá', 'Medellín', 'Cali'],
    'Latitud': [4.711, 6.244, 3.451],
    'Longitud': [-74.072, -75.574, -76.532]
}

# Paso 4: Crear geometrías a partir de las coordenadas (lat, lon)
geometry = [Point(xy) for xy in zip(data['Longitud'], data['Latitud'])]

# Paso 5: Crear el GeoDataFrame
gdf = gpd.GeoDataFrame(data, geometry=geometry, crs="EPSG:4326")

# Paso 6: Mostrar la tabla GeoDataFrame
print("GeoDataFrame:")
print(gdf)

# Paso 7: Mostrar el mapa de puntos
gdf.plot(markersize=100, color='red', edgecolor='black', figsize=(8, 6))
```

5. Spark DataFrame (Procesamiento Distribuido)

Parte de Apache Spark para procesamiento distribuido de datos masivos.

```python
from pyspark.sql import SparkSession

# Crear sesión Spark
spark = SparkSession.builder.appName("Ejemplo Spark").getOrCreate()

# Crear un DataFrame Spark
datos = [('Ana', 25), ('Luis', 30), ('Pedro', 22)]
columnas = ['Nombre', 'Edad']
df = spark.createDataFrame(datos, columnas)

# Mostrar el DataFrame
df.show()
```
